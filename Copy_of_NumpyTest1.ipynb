#Sinus

```python
import numpy as np
import matplotlib.pyplot as plot

frequency = 5_000_000
fs = 250_000_000

nrsamples = np.power(2,13)
samplingtime = np.arange(nrsamples)*(1/fs)

gain  = np.power(2,13)
offset = 0
phase = 0 #np.pi/1

amplitude = gain * np.sin(2*np.pi*frequency*samplingtime + phase) + offset

plotsamples = 100
plot.plot(samplingtime[:plotsamples], amplitude[:plotsamples])
```

```python
float64array = amplitude[::]
print(float64array)
```

```python
int64array = np.int64(float64array[:50])
print(int64array)
```

```python
hex_arr = '\r\n'.join(hex(x) for x in int64array)
print(hex_arr[:64])
```

```python
base_v = np.vectorize(np.base_repr)
padded = np.char.rjust(base_v(int64array, 16), 4, '0')
result = np.char.add('0x', padded)
print(result)
```

```python
for i in result[:50]:
  print(i)
```
